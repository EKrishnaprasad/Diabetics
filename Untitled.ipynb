```python
def univariateAnalysis_numeric(column):
    print("Description of " + column)
    print("----------------------------------------------------------------------------")
    print(dfa[column].describe(),end='\n')
    
    print("\n Skewness of " + column)
    print("----------------------------------------------------------------------------")
    print(np.round(dfa[column].skew(),4),end='\n')
    
    print("\n Histogram and Boxplot of " + column)
    print("----------------------------------------------------------------------------")
    
    fig,axs = plt.subplots(nrows=1,ncols=2, figsize=(14,5))
    print(sns.histplot(dfa[column], kde=True, color='c', ax=axs[0]))
    print(sns.boxplot(x=dfa[column], ax=axs[1],color='r'))
plt.show()
```

```python
num_outliers={}
for col in dfa.columns:
    q1=dfa[col].quantile(0.25)
    q3=dfa[col].quantile(0.75)
    iqr=q3-q1
    outliers=((dfa[col]<(q1-1.5*iqr))|(dfa[col]>(q3+1.5*iqr)))
    num_outliers[col]=outliers.sum()
num_outliers
```
